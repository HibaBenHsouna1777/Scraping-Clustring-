```python
%%writefile spiders/transferspider.py
import scrapy

class TransferspiderSpider(scrapy.Spider):
    name = "transferspider"
    allowed_domains = ["www.transfermarkt.fr"]
    start_urls = ["https://www.transfermarkt.fr/wintertransfers/saisontransfers/statistik?transferfenster=wintertransfers&saison_id=2024"]

    custom_settings = {
        'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
        'DOWNLOAD_DELAY': 2,
        'ROBOTSTXT_OBEY': True
    }

    def parse(self, response):
        for row in response.css('table.items tbody tr'):
            # Extract player name
            player_name = row.css('.hauptlink a::text').get()

            # Extract position from the nested table
            position = row.css('table.inline-table tr:nth-child(2) td::text').get()

            # Extract transfer amount
            transfer_amount = row.css('td.rechts.hauptlink::text').get()

            # Clean and format data
            if player_name and transfer_amount:
                yield {
                    'player_name': player_name.strip(),
                    'position': position.strip() if position else None,
                    'transfer_amount': transfer_amount.strip().replace('\u202f', ' ')  # Replace narrow no-break space
                }

        # Handle pagination
        next_page = response.css('li.naechste-seite a::attr(href)').get()
        if next_page:
            yield response.follow(next_page, callback=self.parse)
```
